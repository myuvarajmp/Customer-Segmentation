### Business Goal:

To segregate credit card holders into follow groups based on their spending history and tenure.
- **Transactos** - Customers who pay the least amount of interest charges and careful with their money.
- **Resolvers** - Customers who use their credit card as a loan. This group is the most lucrative sector for the bank since theu pay 20%+ interest.
- **New Customers** - New customers with low tenure who can be targeted to enroll in other bank services(ex: travel credit card).
- **VIP/PRIME** - Customers with high credit limit/% of full payment. They can be targeted to increase their credit limit/spending.

## Fitting the model

```python
n_clusters=30
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(df_scaled)
    cost.append(kmean.inertia_)
```

```python
plt.plot(cost, 'bx-')
```

```python
# Using Elbow method we can use 7 as our number of clusters
```

```python
kmeans = KMeans(7)
kmeans.fit(df_scaled)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [df.columns])
cluster_centers
```

```python
# In order to understand what these numbers mean, let's perform inverse transformation
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = [df.columns])
cluster_centers
```

- **First Customers cluster (Transactors):** These are customers who pay least amount of interest charges and careful with their money, Cluster with lowest balance (104 dollars) and cash advance (301 dollars), Percentage of full payment = 24%
- **Second customers cluster (revolvers):** who use credit card as a loan (most lucrative sector): highest balance (5000 dollars) and cash advance (~5100 dollars), low purchase frequency, high cash advance frequency (0.5), high cash advance transactions (16) and low percentage of full payment (3%)
- **Third customer cluster (VIP/Prime):** high credit limit 12K dollars and highest percentage of full payment, target for increase credit limit and increase spending habits
- **Fourth customer cluster (low tenure):** these are customers with low tenure (7 years), low balance 

```python
# concatenate the clusters labels to our original dataframe
creditcard_df_cluster = pd.concat([df, pd.DataFrame({'cluster':labels})], axis = 1)
creditcard_df_cluster.head(20)
```

```python
# Plot the histogram of various clusters
for i in df.columns:
  plt.figure(figsize = (35, 5))
  for j in range(7):
    plt.subplot(1,7,j+1)
    cluster = creditcard_df_cluster[creditcard_df_cluster['cluster'] == j]
    cluster[i].hist(bins = 20)
    plt.title('{}    \nCluster {} '.format(i,j))
  
  plt.show()
```

```python
# Obtain the principal components 
pcs = PCA(n_components=2)
principal_comp = pcs.fit_transform(df)
```

```python
# Create a dataframe with the two components
pca_df = pd.DataFrame(data = principal_comp, columns =['pca1','pca2'])
pca_df.head()
```

```python
# Concatenate the clusters labels to the dataframe
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
pca_df.head()
```
